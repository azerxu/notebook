```python
class Cell(object):
    __slots__ = ('cost', 'match', 'qstart', 'tstart')

    def __init__(self, cost=0, match=0, qstart=0, tstart=0):
        self.cost = cost
        self.match = match
        self.qstart = qstart
        self.tstart = tstart

    def __repr__(self):
        return 'cost:{0}, match:{1}, qstart:{2}, tstart:{3}'.format(
            self.cost, self.match, self.qstart, self.tstart)

    def copy(self):
        return Cell(self.cost, self.match, self.qstart, self.tstart)


def locate(query, target, match=1, mismatch=-1, insert=-1, delete=-1):
    """
    query:    query string
    target:   target string
    match:    match score
    mismatch: mismatch score
    insert:   insert score
    delete:   delete score
    """
    m = len(query)
    n = len(target)

    cols = [Cell(0, 0, i, 0) for i in range(m + 1)]

    best_matches = 0
    best_cost = 0
    best_qstart = 0
    best_qend = 0
    best_tstart = 0
    best_tend = 0

    for j in range(1, n + 1):
        temp = cols[0].copy()

        cols[0] = Cell(0, 0, 0, j)

        for i in range(1, m + 1):
            is_match = 1 if (query[i - 1] == target[j - 1]) else 0
            cost_diag = temp.cost + (match if is_match else mismatch)
            cost_insert = cols[i - 1].cost + insert
            cost_delete = cols[i].cost + delete

            if cost_diag >= cost_insert and cost_diag >= cost_delete:
                # match or mismatch
                cost = cost_diag
                matches = temp.match + is_match
                qstart = temp.qstart
                tstart = temp.tstart

            elif cost_insert >= cost_delete:  # insert
                cost = cost_insert
                matches = cols[i - 1].match
                qstart = cols[i - 1].qstart
                tstart = cols[i - 1].tstart

            else:                            # delete
                cost = cost_delete
                matches = cols[i].match
                qstart = cols[i].qstart
                tstart = cols[i].tstart

            temp = cols[i].copy()

            if cols[i].cost < 0:  # reset alignment start if align two low
                cols[i].cost = 0
                cols[i].match = 0
                cols[i].qstart = i
                cols[i].tstart = j
            else:
                cols[i].cost = cost
                cols[i].match = matches
                cols[i].qstart = qstart
                cols[i].tstart = tstart
                if (matches > best_matches or (matches == best_matches and
                                               cost >= best_cost)):
                    best_matches = matches
                    best_cost = cost
                    best_qstart = qstart
                    best_qend = i
                    best_tstart = tstart
                    best_tend = j

    return (best_qstart, best_qend, best_tstart, best_tend,
            best_matches, best_cost)
```
